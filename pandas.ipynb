```python
a = pd.DataFrame({
    'name' : ['John', 'Bob', 'Sam', 'Paul', 'Sarah'],
    'weight' : [80.0,70.4,65.5,45.9,51.2],
    'height' : [170,180,155,143,154],
    'gender' : ['m', 'm', 'm', 'm', 'f']
})

print('Weight')
print(a['weight'])
print('Weight and Height')
print(a[['weight','height']])
print('Height over 170')
print(a[a.height >= 170])
print('Gender m')
print(a[a.gender == 'm'])
print('Height sorted')
print(a.sort_values(by='height'))
print('Weight sorted')
print(a.sort_values(by='weight', ascending=False))
```
