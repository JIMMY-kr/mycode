```python
#계단함수
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def step_function(x):
    return np.array(x > 0, dtype=np.int)

X = np.arange(-5.0, 5.0, 0.1)
y = step_function(X)
plt.plot(X,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
#시그모이드 함수 S자 모양 함수(0 ~ 1.0)
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

X = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(X)
plt.plot(X, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
#하이퍼볼릭 탄젠트, 확장된 시그모이드 함수(-1 ~ 1)
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 12)
y = np.tanh(X)

print(X)
print(y)

plt.plot(X, y, c='r', marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# ReLU 함수

import numpy as np
import matplotlib.pylab as plt

def relu(x):
    return np.maximum(0,x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x,y)
plt.ylim(-1.0, 5.5)
plt.show()
```

```python
#신경망의 연산 : 행렬 사용
x = np.array([1,0.5]) #입력값
#첫번째 은닉층
w1 = np.array([[0.1,0.3,0.5], [0.2,0.4,0.6]])#가중치
b1 = np.array([0.1, 0.2, 0.3])#편향
a1 = np.dot(x, w1) + b1 # 첫번째 은닉층의 합계값
z1 = sigmoid(a1) # 활성화함수
print(a1)
print(z1)

# 두번째 은닉층
w2 = np.array([[0.1,0.4],[0.2,0.5],[0.3,0.6]])
b2 = np.array([0.1, 0.2])
a2 = np.dot(z1, w2)+b2
z2 = sigmoid(a2)
print(a2)
print(z2)

#향등함수 : 항상 성립하는 등식
def identify_function(x):
    return x

w3 = np.array([[0.1,0.3],[0.2,0.4]])
b3 = np.array([0.1, 0.2])
a3 = np.dot(z2, w3)+b3
y = identify_function(a3)
print(y)
```

```python
a = np.array([0.3,2.9,4.0])
exp_a = np.exp(a)
print(exp_a)

sum_exp_a = np.sum(exp_a)
print(sum_exp_a)

y=exp_a / sum_exp_a
print(y)
```
