```python
#수집할 영화
#https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=97816&target=after&page=1

#리뷰 작성자 수
#old_content > h5 > div > strong

#리뷰 점수
#old_content > table > tbody > tr:nth-child(1) > td.title > div > em

import math
from urllib.request import urlopen
from bs4 import BeautifulSoup
#마지막 페이지 번호 계산

html=urlopen('https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=97816&target=after&page=1')
bs = BeautifulSoup(html, 'html.parser')
total = bs.select_one("#old_content > h5 > div > strong")
value = total.text.strip().replace('(','')
pages = int(value)/10
pages = math.ceil(pages)
print('페이지수:', pages)

```

```python
from time import sleep
from urllib.request import urlopen
from bs4 import BeautifulSoup

total_reviews=[]
total_points=[]
for i in range(1, pages+1):
    print(f'page{i}')
    url=(f'https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=97816&target=after&page={i}')
    webpage = urlopen(url)
    source=BeautifulSoup(webpage, 'html.parser')
    reviews=source.select('tbody > tr > td.title')
    for review in reviews:
        total_reviews.append(review)
    points = source.select('td.title > div > em')
    for point in points:
        total_points.append(point.text)
    sleep(0.1)
```

```python
total_reviews2=[]
for review in total_reviews:
    total_reviews2.append(review.text.strip().split('\n\n')[2].replace('\t','')
    .replace('\n', '').replace('네이버', '').replace('신고', ''))

total_reviews2[:5]

items=list(zip(total_reviews2, total_points))
print(len(items))
```

```python
import pandas as pd
df = pd.DataFrame(items, columns=['review', 'point'])
df.head()
```

```python
df.to_csv('c:/data/movie/네이버_호빗.csv', encoding='ms949', index=False)
```
