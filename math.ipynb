```python
def intersection(a,b):
    return list(set(a) & set(b))
```

```python
def number_inter(a,b):
    a1 = str(a)
    b1 = str(b)
    c = []
    c = intersection(a1, b1)
    c.sort()
    return c

a = 1234
b = 3456
c = number_inter(a,b)

print(c)
```

```python
def bmi(weight, height):
    bmi = lambda weight, height: weight / (height) ** 2
    a = bmi(weight, height)
    print('bmi:', f'{a:.2f}')
    if a > 35:
        print('고도비만입니다.')
    elif 30 <= a < 30:
        print('경도비만입니다.')
    elif 23 <= a < 25:
        print('과체중입니다.')
    elif 18.5 <= 23:
        print('정상입니다.')
    else:
        print('저체중입니다.')

bmi(94, 1.84)
```

```python
def my_zip(a,b):
    if len(a) != len(b):
        print('두 입력값의 길이가 다릅니다.')
    else:
        n=len(a)
        c=[]
        for i in range(n):
            c.append((a[i],b[i]))
        return c

a=[1,2,3]
b=['a','b','c']
my_zip(a,b)
```

```python
def saving(money, years, rate):
    result = money*(1+rate/100)**years
    return result
```

```python
money=10000
years=np.arange(0,11)
rate=[3,5,7]
total_money=np.zeros((len(rate),len(years)))
for i in range(len(rate)):
    for j in range(len(years)):
        total_money[i,j] = saving(money, years[j], rate[i])
    plt.plot(years, total_money[i,:], label='{}%'.format(rate[i]))

plt.legend(fontsize = 14)
plt.grid()
plt.xlabel('Years', fontsize=14)
plt.ylabel('Total money', fontsize=14)
```

```python
import pandas as pd
import numpy as np

weight = [22,24,26,30,32,40,65,45,20,29,34,36,36,38,39,48,43,37,33,31,29,39,26,29]
hist = np.zeros(6)
for i in weight:
    if i // 5 == 4:
        hist[0] += 1
    elif i // 5 == 5:
        hist[1] += 1
    elif i // 5 == 6:
        hist[2] += 1
    elif i // 5 == 7:
        hist[3] += 1
    elif i // 5 == 8:
        hist[4] += 1
    elif i // 5 == 9:
        hist[5] += 1

print(hist)
```
