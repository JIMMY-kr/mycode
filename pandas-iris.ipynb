```python
from sklearn import datasets
iris = datasets.load_iris()

print(iris['feature_names'])
```

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(data=np.c_[iris.data, iris.target],
columns=iris['feature_names']+['target']
)

df['target']=df['target'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
df.head()
```

```python
df.to_csv('c:/data/db/iris.csv')
```

```python
X = df.iloc[:, :-1] # :모든 행, :-1 마지막 필드 제외
y = df.iloc[:, [-1]] # 마지막필드만 포함
print(X[:5])
print(y[:5])
print(X.shape)
```

```python
arr = X.mean()
arr
```

```python
import numpy as np
np.savetxt('c:/data/db/avg.csv', \
    arr, fmt='%.1f', delimiter=',')
```

```python
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
engine = create_engine('mysql+mysqldb://web:1234@localhost/myweb', encoding='utf-8')
df.columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
conn = engine.connect()
df.to_sql(name='iris', con=engine, if_exists='replace', index=False)
```

```python
#사이킷런, 패키지 내의 다양한 데이터 라이브러리
from sklearn import datasets
wine=datasets.load_wine()
wine[('feature_names')]
```

```python
import numpy as np
import pandas as pd
#np.c_ >>> combine, X데이터와 y데이터를 합침
df=pd.DataFrame(data=np.c_[wine.data,wine.target],
columns=wine['feature_names']+['target'])
df['target']=df['target'].map({0:'class_0', 1:'class_1', 2:'class_2'}) #종속변수명
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,[-1]]
print(X[:5])
print(y[:5])
print(X.shape)
```

```python
#sql로 병합한 데이터 전송
df.to_csv('c:/data/wine/wine.csv', mode='w', index=False)
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
engine = create_engine('mysql+mysqldb://web:1234@localhost/myweb', encoding='utf-8')
conn = engine.connect()
df.to_sql(name='wine', con=engine, if_exists='replace', index=False)
```
